# Question 3

**Difference between np.array, np.asarray, and np.asanyarray**

*   np.array: Always creates a copy of the input data, regardless of whether it's an array or a list or other sequence-like object.
*   np.asarray: Creates a copy only if the input is not an array already. If the input is already an array, it simply returns a reference to the original array.

*  np.asanyarray: Creates a copy only if the input is not an array, nor a subclass of ndarray. In this way, it's useful if you want to handle subclasses of ndarrays in a generic way.

**Deep Copy vs Shallow Copy**

*   
Shallow copy: A shallow copy creates a new object, but it doesn't copy the nested elements within the array. Instead, it creates references to the original elements. Modifying the original array or its nested elements can also affect the copied array.
*  Deep copy: A deep copy creates a completely new object with independent copies of all the nested elements. If you modify the original array or its nested elements, the copied array will not be affected.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a date range for the time series data
dates = pd.date_range(start='2023-10-26', periods=30, freq='D')

# Generate random temperature and humidity values
temperature = np.random.randint(15, 30, size=30)
humidity = np.random.randint(40, 80, size=30)

# Create the DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Set 'Date' as the index
df = df.set_index('Date')

# Create the plot with dual y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(df.index, df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for Humidity
ax2 = ax1.twinx()  # Create a twin Axes sharing the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(df.index, df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Set the title
plt.title('Temperature and Humidity Over Time')

# Adjust layout to prevent overlapping labels
fig.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Generate x-values for the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

# Calculate the PDF using scipy.stats.norm.pdf
p = norm.pdf(x, loc=np.mean(data), scale=np.std(data))

# Overlay the PDF line plot
plt.plot(x, p, 'r', linewidth=2, label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate two random arrays
x = np.random.rand(100) * 10 - 5  # Generate values between -5 and 5
y = np.random.rand(100) * 10 - 5

# Determine the quadrant for each point
quadrant = np.where(x >= 0, np.where(y >= 0, 1, 4), np.where(y >= 0, 2, 3))

# Create the scatter plot with Seaborn
sns.scatterplot(x=x, y=y, hue=quadrant, palette=['red', 'blue', 'green', 'orange'])

# Add a legend
plt.legend(title='Quadrant')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

```python
# prompt: Using Bokeh, g
# enerate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
# Bar Chart'

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool
import numpy as np
import pandas as pd
from bokeh.transform import linear_cmap

output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=len(categories))

# Create a DataFrame for easier plotting
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a figure
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           toolbar_location=None, tools="")

# Create the bars
bars = p.vbar(x='categories', top='values', width=0.9, source=data,
           legend_field="categories", line_color='white')

# Add color mapping based on values
mapper = linear_cmap(field_name='values', palette='Viridis256', low=min(values), high=max(values))
bars.glyph.fill_color = mapper

# Add hover tooltips to display exact values
hover = HoverTool(tooltips=[('Category', '@categories'), ('Value', '@values')])
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Show the plot
show(p)
```
